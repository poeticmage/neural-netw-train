```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F 
```

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.w1 = torch.tensor(1.43)
        self.b1 = torch.tensor(-0.61)
        self.w2 = torch.tensor(2.63)
        self.b2 = torch.tensor(-0.27)
        self.w3 = torch.tensor(-3.89)
        self.w4 = torch.tensor(1.35)
    def forward(self,x):
        x1=x*self.w1+self.b1
        x2=x*self.w2+self.b2
        y1=F.relu(x1)
        y2=F.relu(x2)
        out=y1*self.w3+y2*self.w4
        return out
```

```python
model=Net()
doses = torch.tensor([0.0, 0.5, 1.0])
model(doses)
```

```python
xi=torch.linspace(start=0,end=1,steps=11)
print(xi)
```

```python
x1t=xi*model.w1+model.b1
x1t
```

```python
y1t=F.relu(x1t)
y1t
```

```python
import seaborn as sns ## seaborn makes it easier to draw nice-looking graphs.
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x=xi,y=y1t,color="pink")
sns.lineplot(x=xi,y=y1t,color="red")
```

```python
x2t=xi*model.w2+model.b2
x1t
```

```python
y2t=F.relu(x2t)
y2t
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x=xi,y=y2t,color="pink")
sns.lineplot(x=xi,y=y2t,color="red")
```

```python
y3=y1t*model.w3
y4=y2t*model.w4
sns.set(style="whitegrid")
sns.scatterplot(x=xi,y=y3,color="lime")
sns.lineplot(x=xi,y=y3,color="green")
sns.set(style="whitegrid")
sns.scatterplot(x=xi,y=y4,color="orange")
sns.lineplot(x=xi,y=y4,color="gold")
```
