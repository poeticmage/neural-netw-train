```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import SGD
import lightning as L
from torch.utils.data import TensorDataset, DataLoader
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
xi=torch.tensor([0.0,0.5,1.0])
yi=torch.tensor([0.0,1.0,0.0])
xt=TensorDataset(xi,yi)
dl=DataLoader(xt)
```

```python
# Creating the neural network
class Net(L.LightningModule):
    def __init__(self):
        super().__init__();
        self.w1=nn.Parameter(torch.tensor(0.6))
        self.b1=nn.Parameter(torch.tensor(0.0))
        self.w2=nn.Parameter(torch.tensor(3.49))
        self.b2=nn.Parameter(torch.tensor(0.0))
        self.w3=nn.Parameter(torch.tensor(-4.11))
        self.w4=nn.Parameter(torch.tensor(2.74))
        self.loss=nn.MSELoss(reduction='sum')

    def forward(self,x):
        x1=x*self.w1+self.b1
        x2=x*self.w2+self.b2
        y1=F.relu(x1)
        y2=F.relu(x2)
        
        final=y1*self.w3+y2*self.w4
        return final

    def configure_optimizers(self):
        return SGD(self.parameters(),lr=0.01)

    def training_step(self,batch,batch_idx):
        print("training")
        inpx,inpy=batch
        outx=self.forward(inpx)
        loss=self.loss(outx,inpy)
        return loss
```

```python
model=Net()
for name,params in model.named_parameters():
    print(name,params.data)
```

```python
outh=model(xi)
outh
# xi
```

```python
inpx=torch.linspace(start=0,end=1,steps=11)
inpx
```

```python
outy=model(inpx)
outy.detach().numpy()
```

```python
sns.scatterplot(x=inpx,y=outy.detach().numpy(),s=200)
sns.lineplot(x=inpx,y=outy.detach().numpy())
sns.scatterplot(x=xi,y=yi,color="red",s=200)
```

```python
model=Net()
print("df")
trainer = L.Trainer(max_epochs=500, # how many times to go through the training data
                    logger=False,
                    enable_checkpointing=False,
                    enable_progress_bar=True)

trainer.fit(model, train_dataloaders=dl)
```
